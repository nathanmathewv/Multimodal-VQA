```python
import pandas as pd
```

```python
full_df = pd.read_csv('Dataset/test_20_withoutyesno.csv')
```

```python
full_df.head()
```

```python
import ast
import shutil
import os

res = []
output_dir = "inference_images"
os.makedirs(output_dir, exist_ok=True)

for i in range(500):
    image_path = full_df['image_path'][i]
    # Copy image to new directory
    shutil.copy(image_path, os.path.join(output_dir, os.path.basename(image_path)))
    # Parse vqa_response string to list
    vqa_list = full_df['vqa_response'][i]
    if isinstance(vqa_list, str):
        try:
            vqa_list = ast.literal_eval(vqa_list)
        except Exception:
            vqa_list = []
    for qa in vqa_list:
        if len(qa) == 2:
            question, answer = qa
            res.append({"image_name": image_path.replace('Dataset/final_dataset/',''), "question": question, "answer": answer})
            print(f"Image Path: {image_path}, Question: {question}, Answer: {answer}")
```

```python
# save res as a csv
res_df = pd.DataFrame(res)
res_df.to_csv('inference_images/vqa_res.csv', index=False)
```
