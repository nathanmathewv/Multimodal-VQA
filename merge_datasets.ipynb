```python
# import all files from Dataset/final_dataset(*)
# put all files of it in a dataframe
import os
import pandas as pd
import glob
import shutil
```

```python
paths = ['Dataset/dataset_nathan', 'Dataset/dataset_sarvesh', 'Dataset/dataset_divyam']
```

```python
#store all files in paths list in a dataframe
all_files = []
for path in paths:
    all_files += glob.glob(os.path.join(path, "*.jpg"))
```

```python
# load merged_image_data_vqa.csv
df = pd.read_csv('Dataset/merged_image_data_vqa.csv')
df.head()
```

```python
output_folder = 'Dataset/final_dataset'
# create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

for index, row in df.iterrows():
    rel_image_path = row['image_path'].replace('Dataset/final_dataset/', '')
    found = False
    for path in paths:
        image_path = os.path.join(path, rel_image_path)
        if image_path in all_files:
            dst = os.path.join(output_folder, os.path.basename(image_path))
            if not os.path.exists(dst):
                shutil.copy(image_path, dst)  # safer than os.rename
                print(f"Copied {image_path} to {dst}")
            else:
                print(f"{dst} already exists")
            found = True
            break
    if not found:
        print(f"{rel_image_path} not found in any source folders")
```

```python
# get length of df
print(f"Length of df: {len(df)}")
# get length of final_dataset
final_dataset_files = glob.glob(os.path.join(output_folder, "*.jpg"))
print(f"Length of final_dataset: {len(final_dataset_files)}")
# check if all files in df are in final_dataset
df_files = df['image_path'].tolist()
df_files = [os.path.join(output_folder, os.path.basename(file)) for file in df_files]
missing_files = [file for file in df_files if file not in final_dataset_files]
print(f"Missing files: {missing_files}")
```
